```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
data = pd.read_csv('csv/time_and_place/범죄발생_장소_20230912140949.csv',encoding='ANSI')
#print(data)
#x범죄 장소
#y 범죄명


#범죄 발생 건수
x_cases = list(data.iloc[0, 2:28])

#2017년
y_2017=list(data.iloc[1, 2:28])#1행,2부터10열가져오기, iloc[행범위, 열범위] 
y_2017=list(map(int,y_2017))#리스트 원소 int로 변경, map(적용시킬 것, 적용시킬 반복 가능한 자료형)
#y_2017 =list((int(data['2017.1'][1]),int(data['2017.2'][1]),int(data['2017.3'][1]),int(data['2017.4'][1]),int(data['2017.5'][1]),int(data['2017.6'][1]),int(data['2017.7'][1]),int(data['2017.8'][1]),int(data['2017.9'][1])))
#2018
y_2018=list(data.iloc[1, 30:56])
y_2018=list(map(int,y_2018))
#2019
y_2019=list(data.iloc[1, 58:84])
y_2019=list(map(int,y_2019))
#2020
y_2020=list(data.iloc[1, 86:112])
y_2020=list(map(int,y_2020))
#2021
y_2021=list(data.iloc[1, 114:140])
y_2021=list(map(int,y_2021))
# 네모 크기
fig, ax = plt.subplots(figsize=(30, 10))
#그래프 표시
ax.plot(x_cases, y_2017,label= '2017년', color='red', linestyle='--', marker='o')
ax.plot(x_cases, y_2018,label= '2018년', color='magenta', linestyle='--', marker='o')
ax.plot(x_cases, y_2019,label= '2019년', color='blue', linestyle='--', marker='o')
ax.plot(x_cases, y_2020,label= '2020년',color='green', linestyle='--', marker='o')
ax.plot(x_cases, y_2021,label= '2021년',color='orange', linestyle='--', marker='o')
#그래프 맨 위 제목
ax.set_title('장소별 범죄 건수')

#x축과 y축
ax.set_xlabel('범죄 발생 장소')
ax.set_ylabel('범죄 건수')

#범례표시

ax.legend(loc = 'best')
plt.xticks(rotation=45)
#표 보여줘
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(10,15))

################################2017######################################
plt.subplot(2,3,1)

x_cases = list(data.iloc[0, 2:28])
y_2017=list(data.iloc[1, 2:28]) # 1행,2부터10열가져오기, iloc[행범위, 열범위] 
y_2017=list(map(int,y_2017)) # 리스트 원소 int로 변경, map(적용시킬 것, 적용시킬 반복 가능한 자료형)

# 그래프 표시

plt.barh( x_cases, y_2017,label= '2017년', color='#ffd7ba')
# 그래프 타이틀
plt.title('2017년도 장소별 범죄 건수')

# x축과 y축
plt.xlabel('범죄 건수')
plt.ylabel('범죄 발생 장소')

###############################2018############################
plt.subplot(2,3,2)#fig사이즈는 인치로함 

x_cases = list(data.iloc[0, 2:28])
y_2018=list(data.iloc[1, 30:56])
y_2018=list(map(int,y_2018))

#그래프 표시
plt.barh( x_cases, y_2018,label= '2018년',  color='#cdb4db')
#그래프 타이틀
plt.title('2018년도 장소별 범죄 건수')

#x축과 y축
plt.xlabel('범죄 건수')

#############################2019##############################
plt.subplot(2,3,3)#fig사이즈는 인치로함
x_cases = list(data.iloc[0, 2:28])
y_2019=list(data.iloc[1, 58:84])
y_2019=list(map(int,y_2019))

#그래프 표시
plt.barh( x_cases, y_2019,label= '2019년',  color='#e76f51')

#그래프 타이틀
plt.title('2019년도 장소별 범죄 건수')

#x축과 y축
plt.xlabel('범죄 건수')

###########################2020#################################
plt.subplot(2,3,4)#fig사이즈는 인치로함
x_cases = list(data.iloc[0, 2:28])
y_2020=list(data.iloc[1, 86:112])
y_2020=list(map(int,y_2020))


#그래프 표시
plt.barh( x_cases, y_2020,label= '2020년',  color='#2a9d8f')
#그래프 타이틀
plt.title('2020년도 장소별 범죄 건수')

#x축과 y축
plt.xlabel('범죄 건수')
plt.ylabel('범죄 발생 장소')

###########################2021#################################
plt.subplot(2,3,5)
x_cases = list(data.iloc[0, 2:28])
y_2021=list(data.iloc[1, 114:140])
y_2021=list(map(int,y_2021))


#그래프 표시
plt.barh( x_cases, y_2021,label= '2021년',  color='green')

#그래프 타이틀
plt.title('2021년도 장소별 범죄 건수')

#x축과 y축
plt.xlabel('범죄 건수')

#표 보여줘
plt.subplots_adjust(left = 0, right = 1.0, bottom = 0.1, top=0.9, wspace = 0.5, hspace =0.3) #subplots_adjust :여백지정
plt.show()
```
