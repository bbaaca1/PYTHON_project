```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

# CSV 파일을 읽어 데이터프레임으로 변환
data = pd.read_csv('csv/population/14세부터인구.csv', encoding='ANSI')
df = pd.DataFrame(data)

# 연령대별로 총 인구를 저장할 리스트 초기화
from_fourteen = []
from_nineteen = []
from_twenty_one = []
from_thrity_one = []
from_fourty_one = []
from_fifty_one = []
from_sixty_one = []
from_seventy_one = []

# 데이터프레임의 열 범위 설정
start_index = [1, 6, 8, 18, 28, 38, 48, 58]
end_index = [6, 8, 18, 28, 38, 48, 58, 68]


result_list = []
# 연도별, 연령대별로 총 인구 합계 계산
for i in range(2, 9):
    for j in range(0, 8):
        selected_column = df.iloc[start_index[j]:end_index[j], i].astype(int)
        total = selected_column.sum()
        if j == 0:
            from_fourteen.append(total)
        elif j == 1:
            from_nineteen.append(total)
        elif j == 2:
            from_twenty_one.append(total)
        elif j == 3:
            from_thrity_one.append(total)
        elif j == 4:
            from_fourty_one.append(total)
        elif j == 5:
            from_fifty_one.append(total)
        elif j == 6:
            from_sixty_one.append(total)
        elif j == 7:
            from_seventy_one.append(total)

    
for i in range(0, 7) :
    total = 0
    total += from_fourteen[i]
    total += from_nineteen[i]
    total += from_twenty_one[i]
    total += from_thrity_one[i]
    total += from_fourty_one[i]
    total += from_fifty_one[i]
    total += from_sixty_one[i]
    total += from_seventy_one[i]
    result_list.append(total)
            

# 그래프의 x축 레이블 정의
years = ['2015년', '2016년', '2017년', '2018년', '2019년', '2020년', '2021년']

# y축에 표시할 눈금 값과 레이블 정의
population = range(1000000, 9000001, 1000000)
population_labels = ['1백만', '2백만', '3백만', '4백만', '5백만', '6백만', '7백만', '8백만', '9백만']

# 연령대별 총 인구 데이터를 딕셔너리로 정리
data_dict = {
    '14세~18세': from_fourteen,
    '19세~20세': from_nineteen,
    '21세~30세': from_twenty_one,
    '31세~40세': from_thrity_one,
    '41세~50세': from_fourty_one,
    '51세~60세': from_fifty_one,
    '61세~70세': from_sixty_one,
    '71세이상': from_seventy_one
}


# 다양한 마커 스타일 정의 (점모양, 선 스타일, 색상)
marker_styles = ['o', 's', 'D', '^', 'v', '>', '<', 'p']

# 그래프 생성 및 크기 설정
plt.figure(figsize=(10, 6))
for i, (label, values) in enumerate(data_dict.items()):
    marker_style = marker_styles[i % len(marker_styles)]  # 마커 스타일 순환
    plt.plot(years, values, label=label, marker=marker_style)

# 그래프 제목과 레이블 설정
plt.title("연도별 인구수")
plt.xlabel("연도")
plt.ylabel("인구수")

# 범례 위치 조정 (그래프 오른쪽 상단에 위치)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# y축 눈금과 레이블 설정
plt.yticks(population, population_labels)

# 그래프 출력
plt.show()
```

```python
# CSV 파일을 읽어 데이터프레임으로 변환
data = pd.read_csv('csv/population/범죄자_연도별연령.csv', encoding='ANSI')
df = pd.DataFrame(data)

# 연령대별로 총 범죄자수를 저장할 리스트 초기화
juvenile_offender = []
crime_nineteen = []
crime_twenty_one = []
crime_thrity_one = []
crime_fourty_one = []
crime_fifty_one = []
crime_sixty_one = []
crime_seventy_one = []

# 데이터프레임의 열 범위 설정
start_index = [2, 7, 9, 15, 17, 19, 21, 22]
end_index = [7, 9, 15, 17, 19, 21, 22, 23]


crime_result_list = []
# 연도별, 연령대별로 총 범죄자수 합계 계산
for i in range(4, 11):
    for j in range(0, 8):
        selected_column = df.iloc[start_index[j]:end_index[j], i].fillna(0).astype(int)
        total = selected_column.sum()
        if j == 0:
            juvenile_offender.append(total)
        elif j == 1:
            crime_nineteen.append(total)
        elif j == 2:
            crime_twenty_one.append(total)
        elif j == 3:
            crime_thrity_one.append(total)
        elif j == 4:
            crime_fourty_one.append(total)
        elif j == 5:
            crime_fifty_one.append(total)
        elif j == 6:
            crime_sixty_one.append(total)
        elif j == 7:
            crime_seventy_one.append(total)

    
for i in range(0, 7) :
    total = 0
    total += juvenile_offender[i]
    total += crime_nineteen[i]
    total += crime_twenty_one[i]
    total += crime_thrity_one[i]
    total += crime_fourty_one[i]
    total += crime_fifty_one[i]
    total += crime_sixty_one[i]
    total += crime_seventy_one[i]
    crime_result_list.append(total)
    
# 그래프의 x축 레이블 정의
years = ['2015년', '2016년', '2017년', '2018년', '2019년', '2020년', '2021년']

# 연령대별 총 인구 데이터를 딕셔너리로 정리
data_dict = {
    '14세~18세': juvenile_offender,
    '19세~20세': crime_nineteen,
    '21세~30세': crime_twenty_one,
    '31세~40세': crime_thrity_one,
    '41세~50세': crime_fourty_one,
    '51세~60세': crime_fifty_one,
    '61세~70세': crime_sixty_one,
    '71세이상': crime_seventy_one
}

# 그래프 생성 및 크기 설정
plt.figure(figsize=(10, 6))

# 연령대별로 막대 그래프 그리기
num_bars = len(years)
bar_width = 0.1  # 각 막대의 너비 설정
index = np.arange(num_bars)  # 각 그래프의 x 좌표 설정

for i, (label, values) in enumerate(data_dict.items()):
    plt.bar(index + (i * bar_width), values, bar_width, label=label)

# 그래프 제목과 레이블 설정
plt.title("연도별 범죄자수")
plt.xlabel("연도")
plt.ylabel("범죄자수")

# 범례 위치 조정 (그래프 오른쪽 상단에 위치)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# x 축 눈금과 레이블 설정
plt.xticks(index + (num_bars * bar_width / 2), years)

# 그래프 출력
plt.show()
```
